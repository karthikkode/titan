```python
import os
from pyspark.sql import SparkSession

# 1. Initialize Spark Session connected to Cluster
spark_master = os.environ.get("SPARK_MASTER", "spark://spark-master:7077")

print(f"Connecting to {spark_master}...")

spark = SparkSession.builder \
    .appName("Jupyter-Interactive-Test") \
    .master(spark_master) \
    .config("spark.driver.host", "titan-jupyter") \
    .config("spark.driver.bindAddress", "0.0.0.0") \
    .config("spark.executor.memory", "512m") \
    .getOrCreate()

print("Connected!")
spark
```

```python
# 2. Test Execution on Cluster
data = [("Alice", 1), ("Bob", 2), ("Charlie", 3)]
df = spark.createDataFrame(data, ["Name", "Value"])

print("Running simple count action on cluster...")
print(f"Row count: {df.count()}")
df.show()
```
